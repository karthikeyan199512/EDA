### Most successful team at Home

```python
most_succ = match[match['Team_Name_Id']==match['Match_Winner_Id']]['Team_Name_Id'].value_counts().reset_index() #Team_Name_Id indicates that it is the home team
```

```python
succ_home = pd.merge(most_succ,team,left_on='index',right_on='Team_Id')[['Team_Name','Team_Name_Id']].rename(columns={'Team_Name_Id':'Home_wins'})
```

```python
succ_home
```

```python
# To calculate home win percentage
home_per = (match[match['Team_Name_Id']==match['Match_Winner_Id']]['Team_Name_Id'].value_counts()/match['Team_Name_Id'].value_counts())*100
```

```python
# To calculate away win percentage
away_per = (match[match['Opponent_Team_Id']==match['Match_Winner_Id']]['Opponent_Team_Id'].value_counts()/match['Opponent_Team_Id'].value_counts())*100
```

```python
# To calculate total win percentage
tot_per = (match[match['Team_Name_Id']==match['Match_Winner_Id']]['Team_Name_Id'].value_counts()+match[match['Opponent_Team_Id']==match['Match_Winner_Id']]['Opponent_Team_Id'].value_counts())/(match['Team_Name_Id'].value_counts()+match['Opponent_Team_Id'].value_counts())*100
```

```python
df1 = pd.merge(home_per,team,left_index=True,right_on='Team_Id')[['Team_Name','Team_Name_Id']].rename(columns={'Team_Name_Id':'Home_win_%'})
```

```python
df2 = pd.merge(away_per,team,left_index=True,right_on='Team_Id')[['Team_Name','Opponent_Team_Id']].rename(columns={'Team_Name_Id':'Away_win_%'})
```

```python
df3 = ((match[match['Team_Name_Id']==match['Match_Winner_Id']]['Team_Name_Id'].value_counts()+match[match['Opponent_Team_Id']==match['Match_Winner_Id']]['Opponent_Team_Id'].value_counts())/(match['Team_Name_Id'].value_counts()+match['Opponent_Team_Id'].value_counts())*100).reset_index()
```

```python
temp_merge1 = pd.merge(df1,team,left_on='Team_Name',right_on='Team_Name')
```

```python
temp_merge2 = pd.merge(temp_merge1,df2,left_on='Team_Name',right_on='Team_Name')
```

```python
final_merge = round(pd.merge(temp_merge2,df3,left_index=True,right_index=True)[['Team_Name','Home_win_%','Opponent_Team_Id',0]].rename(columns={'Opponent_Team_Id':'Away_win_%',0:'Tot_win_%'}),2)
```

```python
final_merge
```

### Top 10 run scorers in IPL History

```python
# reading a new file which contains ball by ball data
ball = pd.read_excel('Ball_by_Ball.xlsx')
```

```python
ball.head()
```

```python
bats_highest = ball.groupby('Striker_Id').agg('sum')['Batsman_Scored'].sort_values(ascending=False)
```

```python
# Reading data containg the Player names
player = pd.read_excel('Player.xlsx')
```

```python
# Merging the Player names
bats_high = pd.merge(bats_highest,player,left_index=True,right_on='Player_Id').head(10)[['Player_Name','Batsman_Scored']].reset_index(drop=True).rename(columns={'Batsman_Scored':'Runs_Scored'})
```

```python
bats_high
```

### Top 10 wicket takers in IPL History

```python
ball['Dissimal_Type'].unique()
```

```python
bowl = ball[(ball['Dissimal_Type']=='caught') | (ball['Dissimal_Type']=='bowled') | (ball['Dissimal_Type']=='lbw') | (ball['Dissimal_Type']=='stumped') | (ball['Dissimal_Type']=='caught and bowled') | (ball['Dissimal_Type']=='hit wicket')]
```

```python
bowl_highest = bowl.groupby('Bowler_Id').agg('count')['Dissimal_Type'].sort_values(ascending=False)
```

```python
bowl_high = pd.merge(bowl_highest,player,left_index=True,right_on='Player_Id').head(10)[['Player_Name','Dissimal_Type']].reset_index(drop=True).rename(columns={'Dissimal_Type':'Wickets_Taken'})
```

```python
bowl_high
```

### Top 10 highest individual scores in IPL History

```python
runs_high = ball.groupby(['Match_Id','Striker_Id'])['Batsman_Scored'].agg('sum').sort_values(ascending=False).head(10)
```

```python
final_scores_high = pd.merge(runs_high,player,left_on='Striker_Id',right_on='Player_Id')[['Player_Name','Batsman_Scored']].sort_values(by='Batsman_Scored',ascending=False).reset_index(drop=True).rename(columns={'Batsman_Scored':'Highest_Score'})
```

```python
final_scores_high
```

### Top 5 century makers in IPL history

```python
top_cen = ball.groupby(['Match_Id','Striker_Id'])['Batsman_Scored'].agg('sum').sort_values(ascending=False)
```

```python
top_cen1 = pd.merge(top_cen,player,left_on='Striker_Id',right_on='Player_Id')[['Player_Name','Batsman_Scored']]
```

```python
top_cen2 = top_cen1[top_cen1['Batsman_Scored'] >= 100]
```

```python
top_cen2.groupby('Player_Name')['Batsman_Scored'].agg('count').sort_values(ascending=False).head(5).reset_index().rename(columns={'Batsman_Scored':'Century_Scored'})
```

### Top 10 half-century makers in IPL history

```python
final_50 = top_cen1[top_cen1['Batsman_Scored'] >= 50].groupby('Player_Name')['Batsman_Scored'].agg('count').sort_values(ascending=False).head(10)
```

```python
final_50.reset_index().rename(columns={'Batsman_Scored':'Number_of_fifties'})
```
